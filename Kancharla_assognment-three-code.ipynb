```python
#Question 1.
import re
from collections import Counter
import nltk
from nltk.corpus import stopwords

text = "Harry lay in his dark cupboard much later, wishing he had a watch. He didn't know what time it was and he couldn't be sure the Dursleys were asleep yet. Until they were, he couldn't risk sneaking to the kitchen for some food. He'd lived with the Dursleys almost ten years, ten miserable years, as long as he could remember, ever since he'd been a baby and his parents had died in that car crash. He couldn't remember being in the car when his parents had died. Sometimes, when he strained his memory during long hours in his cupboard, he came up with a strange vision: a blinding flash of green light and a burning pain on his forehead. This, he supposed, was the crash, though he couldn't imagine where all the green light came from. He couldn't remember his parents at all. His aunt and uncle never spoke about them, and of course he was forbidden to ask questions. There were no photographs of them in the house."
#(1)
vowel_words = re.findall(r'\b[aeiou]\w+', text, flags=re.IGNORECASE)
vowel_freq = Counter(vowel_words)
print(sorted(vowel_freq.items(), key=lambda x: x[1], reverse=True))
```

```python
#(2)
apostrophe_words = re.findall(r"\b\w+'\w+\b", text)
print(apostrophe_words)
```

```python
import csv
import statistics
import os

# Define a function to calculate mean, median, and standard deviation
def calculate_stats(data):
    mean = statistics.mean(data)
    median = statistics.median(data)
    stdev = statistics.stdev(data)
    return mean, median, stdev

# Define a function to read the data from a file
def read_data(filename):
    high_temps = []
    low_temps = []
    precipitations = []
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row
        for row in reader:
            high_temps.append(float(row[1]))
            low_temps.append(float(row[2]))
            precipitations.append(float(row[3]))
    return high_temps, low_temps, precipitations

# Define column names
column_names = ["Filename", "Mean", "Median", "Stdev"]

# Print column names for each statistic
for stat in ["High Temp", "Low Temp", "Precip"]:
    print(f"\n{stat}:\n")
    for name in column_names:
        print(f"{name.ljust(20)}", end='')
    print()

    # Call the functions to calculate the stats for each file
    files = ['C:/Users/Kancharla/Desktop/BS/19th_ass/2011-Jan-June.csv', 'C:/Users/Kancharla/Desktop/BS/19th_ass/2018-Jan-June.csv']
    for filename in files:
        base_filename = os.path.basename(filename)
        high_temps, low_temps, precipitations = read_data(filename)
        mean, median, stdev = calculate_stats(high_temps if stat == "High Temp" else low_temps if stat == "Low Temp" else precipitations)
        row_values = [base_filename, f"{mean:.2f}", f"{median:.2f}", f"{stdev:.2f}"]
        for value in row_values:
            print(f"{value.ljust(20)}", end='')
        print()
```

```python
import random

class Student:
    def __init__(self, first_name, last_name, middle_name=''):
        self.first_name = first_name
        self.middle_name = middle_name
        self.last_name = last_name
        self.euid = ''
        self.gpa = 0
        self.classes_taken = []

    def assign_euid(self):
        euid_letters = self.first_name[:1].lower() + self.middle_name[:1].lower() + self.last_name[:1].lower()
        euid_numbers = str(random.randint(1000, 9999))
        self.euid = euid_letters + euid_numbers
       
    def register(self):
        class_number = input("Enter class number: ")
        semester = input("Enter semester: ")
        grade = input("Enter grade: ")
        self.classes_taken.append((class_number, semester, grade))

    def calculate_gpa(self):
        total_credits = 0
        total_grade_points = 0
        for course in self.classes_taken:
            grade_points = self.convert_grade_to_points(course[2])
            credits = self.get_credits(course[0])
            total_grade_points += grade_points * credits
            total_credits += credits
        self.gpa = round(total_grade_points / total_credits, 2)

    def convert_grade_to_points(self, grade):
        if grade == 'A':
            return 4.0
        elif grade == 'A-':
            return 3.7
        elif grade == 'B+':
            return 3.3
        elif grade == 'B':
            return 3.0
        elif grade == 'B-':
            return 2.7
        elif grade == 'C+':
            return 2.3
        elif grade == 'C':
            return 2.0
        elif grade == 'C-':
            return 1.7
        elif grade == 'D+':
            return 1.3
        elif grade == 'D':
            return 1.0
        else:
            return 0

    def get_credits(self, class_number):
        # This function would typically look up the number of credits for a given class
        # number in a database, but for simplicity we'll just return a hardcoded value.
        return 3

# Prompt the user to enter information for three students
students = []
for i in range(3):
    first_name = input("Enter first name for student {}: ".format(i+1))
    middle_name = input("Enter middle name for student {} (press enter if none): ".format(i+1))
    last_name = input("Enter last name for student {}: ".format(i+1))
    student = Student(first_name, last_name, middle_name)
    students.append(student)

# Assign euids to the students
for student in students:
    student.assign_euid()
    print("\n{} {} (EUID: {})".format(student.first_name, student.last_name, student.euid))

# Register classes for the students
for student in students:
    print("\nRegistering classes for {} {}".format(student.first_name, student.last_name))
    while len(student.classes_taken) < 2:
        student.register()

# Calculate GPAs for the students
for student in students:
    student.calculate_gpa()

# Print out the names and GPAs of the students
for student in students:
    print("\n{} {}".format(student.first_name, student.last_name))
    print("GPA: {}".format(student.gpa))
```

```python
import random

def quiz_introduction():
    print("************************")
    print("** A Sample Math Quiz **")
    print("************************")

def quiz_questions(index):
    if index == 1:
        num1 = random.randint(1, 100)
        num2 = random.randint(1, 100)
        problem = f"{num1} + {num2}"
        solution = num1 + num2
    elif index == 2:
        num1 = random.randint(1, 100)
        num2 = random.randint(1, 100)
        problem = f"{num1} - {num2}"
        solution = num1 - num2
    elif index == 3:
        num1 = random.randint(1, 10)
        num2 = random.randint(1, 10)
        problem = f"{num1} * {num2}"
        solution = num1 * num2
    elif index == 4:
        num1 = random.randint(1, 100)
        num2 = random.randint(1, 10)
        problem = f"{num1} // {num2}"
        solution = num1 // num2
    else:
        print("Invalid index.")
        return 0
    user_solution = get_user_input_solution(problem)
    count = solution_checking(user_solution, solution, 0)
    return count

def get_user_question_choice():
    print("Please select a topic to answer:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Integer Division")
    print("5. Exit")
    choice = input("Enter your choice: ")
    return int(choice)

def get_user_input_solution(problem):
    print("Enter your answer:")
    print(problem, end="")
    result = int(input(" = "))
    return result

def solution_checking(user_solution, solution, count):
    if user_solution == solution:
        count += 1
        print("Correct!")
    else:
        print(f"Incorrect. The correct answer is {solution}.")
    return count

def section_separator():
    print("------------------------")

def show_result(total, correct):
    score = (correct / total) * 100
    print(f"You answered {total} questions with {correct} correct.")
    print(f"Your score is {score:.1f}%. Thank you.")

def main():
    quiz_introduction()
    section_separator()
    option = get_user_question_choice()
    total = 0
    correct = 0
    while option != 5:
        total += 1
        correct = quiz_questions(option)
        option = get_user_question_choice()
    print("Exiting the quiz.")
    section_separator()
    show_result(total, correct)

main()
```
